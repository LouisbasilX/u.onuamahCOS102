```python
def getNumberInput(positive, text):
  while True:
    try:
        # 1. Get input and attempt conversion
        user_input = float(input(f"Enter a {text}: "))
        if (not positive and user_input != 0) or (positive and user_input > 0):
         return user_input
        else:
         print(f"You have not entered a {text}! Please Enter a {text}")  
    except ValueError:
        # 3. This runs only if int() failed
        print("Invalid input. Please enter real numbers only (e.g., 123, 1.2, 2.3).")
payments = getNumberInput(True, "Positive Payment")
compoundTimes = getNumberInput(True, "Positive Compound period(per year)")
time          = getNumberInput(True, "Positive TIme(years)")
rate_ratio      = getNumberInput(False, "Non zero Rate")/compoundTimes/100

amount = payments * ((1.0 + rate_ratio) ** (compoundTimes * time) - 1.0)/ rate_ratio
# Total money you actually paid out of pocket
total_principal = payments * compoundTimes * time

# The "Annuity" value (Interest/Gain)
annuity_gain = amount - total_principal

print(f"Total Amount: {amount:.2f}")
print(f"Total Principal Paid: {total_principal:.2f}")
print(f"Annuity Interest Gained: {annuity_gain:.2f}") # Here it is!
        
```
