```python
def getNumberInput(positive, text):
  while True:
    try:
        # 1. Get input and attempt conversion
        user_input = float(input(f"Enter a {text}: "))
        if not positive or (positive and user_input > 0):
         return user_input
        else:
         print(f"You have entered a non {text}! Please Enter a {text}")  
    except ValueError:
        # 3. This runs only if int() failed
        print("Invalid input. Please enter real numbers only (e.g., 123, 1.2, 2.3).")
principal = getNumberInput(True, "Positive Principal")
rate      = getNumberInput(False, "Rate")
compoundTimes = getNumberInput(True, "Positive Compound period(per year)")
time          = getNumberInput(True, "Positive TIme(years)")

amount = principal * (1.0 + rate / compoundTimes / 100) ** (compoundTimes * time)
compoundInterest = amount - principal
print(f"Your compound interest is {compoundInterest} and your amount is {amount}")
        
```
