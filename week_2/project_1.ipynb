START

    // 1. Initialization
    SET itemList = []
    SET total_price = 0.00
    SET shipping_info_dict = {}

    // 2. Item Entry Loop
    WHILE (customer_has_more_items) DO
        INPUT barcode
        
        IF (barcode EXISTS in database) THEN
            item = FETCH_ITEM(barcode)
            APPEND item TO itemList
            total_price = total_price + item.price
            
            DISPLAY item.name, item.price
            DISPLAY "Subtotal: ", total_price
        ELSE
            DISPLAY "ERROR: Item not found."
        END IF
    END WHILE

    // 3. Shipping Info Entry Loop
    WHILE (customer_has_shipping_data) DO
        INPUT label, value
        SET shipping_info_dict[label] = value
    END WHILE

    // 4. Function Definition
    FUNCTION print_summary(data_dict, total)
        DISPLAY "--- SHIPPING DETAILS ---"
        FOR EACH entry IN data_dict DO
            DISPLAY entry.label + ": " + entry.value
        END FOR
        
        DISPLAY "------------------------"
        DISPLAY "GRAND TOTAL: " + total
    END FUNCTION

    // 5. Execution
    CALL print_summary(shipping_info_dict, total_price)

END
